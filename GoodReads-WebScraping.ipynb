```python
import pandas as pd
import numpy as np
from selenium import webdriver
import random
import time 

import pandas as pd
import numpy as np
from selenium import webdriver
import random
import time

def scrape_book(i):
    """Scrape all info from a singular book."""
    
    try:
        # Click into the book
        rank = i
        
        # Access the book link on the main page
        book_button = driver.find_element('xpath', f'//*[@id="all_votes"]/table/tbody/tr[{i}]/td[3]/a/span')
        book_button.click()
       
        # Check for pop-up ad
        time.sleep(random.uniform(1.9, 2.4))
        try:
            ad_button = driver.find_element('xpath', '/html/body/div[3]/div/div[1]/div/div/button')
            ad_button.click()
            time.sleep(random.uniform(1.9, 2.4))
    
        except:
            print('no ad') 
       
        # Scrape info
        book_info = []
        title = driver.find_element('xpath', '//*[@id="__next"]/div[2]/main/div[1]/div[2]/div[2]/div[1]/div[1]/h1').text
        
        # Extract publication date
        try:
            try:
                pub_date = driver.find_element('xpath', '//*[@id="__next"]/div[2]/main/div[1]/div[2]/div[2]/div[2]/div[6]/div/span[1]/span/div/p[2]').text
            except:
                pub_date = driver.find_element('xpath', '//*[@id="__next"]/div[2]/main/div[1]/div[2]/div[2]/div[2]/div[6]/div/span[1]/span/div/p').text
        except:
            pub_date = driver.find_element('xpath', '//*[@id="__next"]/div[2]/main/div[1]/div[2]/div[2]/div[2]/div[7]/div/span[1]/span/div/p[2]').text
        # Extract other book details
        author = driver.find_element('xpath', '//*[@id="__next"]/div[2]/main/div[1]/div[2]/div[2]/div[2]/div[1]/h3/div/span[1]/a/span[1]').text
        avg_rating = driver.find_element('xpath', '//*[@id="__next"]/div[2]/main/div[1]/div[2]/div[2]/div[2]/div[2]/a/div[1]/div').text
        num_ratings = driver.find_element('xpath', '//*[@id="__next"]/div[2]/main/div[1]/div[2]/div[2]/div[2]/div[2]/a/div[2]/div/span[1]').text
        try:
           pages = driver.find_element('xpath', '//*[@id="__next"]/div[2]/main/div[1]/div[2]/div[2]/div[2]/div[6]/div/span[1]/span/div/p[1]').text
        except:
            pages = driver.find_element('xpath', '//*[@id="__next"]/div[2]/main/div[1]/div[2]/div[2]/div[2]/div[7]/div/span[1]/span/div/p[1]').text
        
        # Optional fields
        try:
            genre = driver.find_element('xpath', '//*[@id="__next"]/div[2]/main/div[1]/div[2]/div[2]/div[2]/div[5]/ul/span[1]/span[2]/a/span').text
        except:
            genre = None
    
        try:
            num_reading = driver.find_element('xpath', '//*[@id="__next"]/div[2]/main/div[1]/div[2]/div[2]/div[2]/div[7]/div/div[1]/span/div[2]/div').text
        except:
            num_reading = driver.find_element('xpath', '//*[@id="__next"]/div[2]/main/div[1]/div[2]/div[2]/div[2]/div[8]/div/div[1]/span/div[2]/div').text
        try:
            num_want_read = driver.find_element('xpath', '//*[@id="__next"]/div[2]/main/div[1]/div[2]/div[2]/div[2]/div[7]/div/div[2]/span/div[2]/div').text
        except:
            num_want_read = driver.find_element('xpath', '//*[@id="__next"]/div[2]/main/div[1]/div[2]/div[2]/div[2]/div[8]/div/div[2]/span/div[2]/div').text
        # Language (if available)
        try:
            language = driver.find_element('xpath', '//*[@id="__next"]/div[2]/main/div[1]/div[2]/div[2]/div[2]/div[6]/div/span[2]/div[1]/span/div/dl/div[5]/dd/div/div[1]').text
        except:
            language = None
    
        # Back to book list page
        time.sleep(random.uniform(0.7,1.2))
        driver.back()
        
        if pub_date:  # return book info if publication date is available
            book_info.append([title, pub_date, author, genre, avg_rating, num_ratings, pages, rank, language, num_reading, num_want_read])
        return book_info
    
    except:
        driver.back()  # Ensure the browser navigates back even if an error occurs
        return None  # Return None or an empty list to indicate failure





def scrape_page(books_per_page):
    """Scrape a page of books."""
    books = []
    for i in range(1, books_per_page + 1):
        print(f"Scraping book {i}")
        book = scrape_book(i)
        if book:  # Only add to the list if the book scrape was successful
            books.extend(book)
        else:
            print(f"Book {i} failed")
    return books


def scrape_all_pages(url, pages, books_per_page):
    """Scrape all pages of books."""
    books = []
    for i in range(1, pages + 1):
        page_url = f"{url}{i}"
        driver.get(page_url)
        print(f'Scraping page {i}')
        books.extend(scrape_page(books_per_page))
        time.sleep(random.uniform(1.9, 2.4))

    # Save to CSV
    books_df = pd.DataFrame(books, columns=['Title', 'Publication Date', 'Author', 'Genre', 'Average Rating', 'Number of Ratings', 
                                            'Pages', 'Rank','Language', 'Currently Reading', 'Want to Read'])
    books_df.to_csv('task1.csv')
    return books_df



```

```python
def convert_abbrev_num(num_str):
    # Remove commas for clean conversion
    num_str = num_str.replace(',', '')
    
    # Check if the last character is a digit and directly convert to integer if true
    if num_str[-1].isdigit():
        return int(num_str)
    
    # Extract the numeric part and the suffix
    number = num_str[:-1]
    suffix = num_str[-1].lower()
    
    # Determine the multiplier based on the suffix
    multiplier = 1  # Default is no multiplication
    if suffix == 'k':
        multiplier = 1000
    elif suffix == 'm':
        multiplier = 1000000
    elif suffix == 'b':
        multiplier = 1000000000
    else:
        raise ValueError(f"Unsupported suffix '{suffix}' in input.")
    
    # Return the numeric part multiplied by the determined multiplier
    return int(float(number) * multiplier)
```
